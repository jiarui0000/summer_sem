```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt

def cost_function(theta_value):
    X = 10
    s = 5
    h = 2
    p = 8
    K = 10
    c = 0
    demand = np.random.exponential(1/theta_value, 1000)
    for i in range(1000):
        if(demand[i] <= X - s):
            c += h * (X - demand[i])
            X -= demand[i]
        elif(demand[i] > X - s and demand[i] <= X):
            c += h * (X - demand[i]) + K
            X = 10
        else:
            c += p * (demand[i] - X) + K
            X = 10
    return c / 1000

def gradient(theta_value, c):
    return c - 37

# m = number of iterations
# ep = epsilon fixed step size
    
def stochasticapproximation(th0, m, ep, eta):
    theta = np.zeros(m)
    cf = np.zeros(m)
        
    theta[0] = th0
    for i in range(m-1):
        cf[i]= cost_function(theta[i])
        
        # gradient descent
        
        theta[i+1] = max(0.00001, (theta[i] - ep * gradient(theta[i], cf[i]) / (i + 1)))
        
        # FD descent
        
        #etai = (   eta/(2*(i+1))  )**2.0
        #cost_plus = cost_function(theta[i]+etai )
        #cost_minus = cost_function(theta[i]-etai )
        #cost_diff = ( cost_plus - cost_minus)/ ( 2.0*etai )
        #theta[i+1] = theta[i] - ep*cost_diff/(i+1)   
    
    cf[m-1]= cost_function(theta[m-1])
    
    plt.plot(np.arange(m),theta)
    plt.legend(['theta']) 
    plt.show()

    plt.plot(np.arange(m),cf)
    plt.legend(['cost function']) 
    plt.show()


    print("Final values  theta & cost", theta[m-1], cost_function(theta[m-1]))    
    
    return theta
```

```python
def main():
    th0= 1.0
    m  = 100 # iterations
    ep = 0.075  
    eta = 1.0
    
    stochasticapproximation(th0,m,ep,eta)
    
if __name__ == '__main__':
    main()
```
